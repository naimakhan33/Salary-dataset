```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv(r"C:\Users\Ahsan\Downloads\archive.zip")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset["Salary"]
```

```python
x.ndim
```

```python
dataset.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=46)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# y1=m1x1+m2x2+c
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
x_test = x_test.rename(columns={'Unnamed: 0': 'Index'})

y_pnd = (-1705.95309681 * x_test['Index'] 
         + 5692.42333133 * x_test['YearsExperience'] 
         + 80395.36981358992)

print(y_pnd)


```

```python
plt.figure(figsize=(8, 5))
plt.plot(x_test['YearsExperience'], y_pnd, 'bo-', label='Predicted Salary')
plt.plot(x_test['YearsExperience'], y_test, 'r*-', label='Actual Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Actual vs Predicted Salary')
plt.legend()
plt.grid(True)
plt.show()



```
